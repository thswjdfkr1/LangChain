```python
from google.colab import userdata
open_ai_key = userdata.get('Openai')
claude_ai_key = userdata.get('claude')
gemini_ai_key = userdata.get('gemini')
```

## open ai

```python
from langchain_openai import ChatOpenAI
```

```python
model = ChatOpenAI(api_key = open_ai_key, model_name='gpt-4o')
```

```python
resp = model.invoke("안녕")
```

```python
resp
```

```python
resp.content
```

## gemini

```python
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
model = ChatGoogleGenerativeAI(api_key = gemini_ai_key, model="gemini-1.5-pro")
resp = model.invoke("안녕")
print(resp.content)
```

## Anthropic Claude

```python
from langchain_anthropic import ChatAnthropic
```

```python
model = ChatAnthropic(api_key = claude_ai_key, model="claude-3-haiku-20240307")
resp = model.invoke("안녕")
print(resp.content)
```

```python
resp = model.invoke("html, js, tailwindcss로 loginform 만들어줘")
```

```python
from IPython.display import Markdown
import textwrap

def to_markdown(text):
  text = text.replace('•', '  *')
  return Markdown(text.wrap(text, '> ',predicate=lambda _:True))
```

```python
to_markdown(resp.content)
```
